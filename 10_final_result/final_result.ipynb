# Voorbereiding

```python
from tensorflow import keras
from keras import layers
import numpy as np
import matplotlib.pyplot as plt
import io
from PIL import Image
import os
from google.colab import drive
import matplotlib.image as mpimg
```

# Classifying two painters

```python
model = keras.models.load_model("data/model_two_painters.h5")
```

```python
from ipywidgets import FileUpload
uploader = FileUpload(multiple = True)
uploader
```

```python
if os.path.exists("images"):
  os.system("rm -rf "+"images")

os.mkdir('images')

imagePaths = []
i=0
for name, file_info in uploader.value.items():
  print(name)
  img = Image.open(io.BytesIO(file_info['content']))
  path = "images/" + str(i) + ".jpg"
  img.save(path)
  imagePaths.append(path)
  i+=1
```

```python
predictions = [0] * len(imagePaths)
for i in range(len(imagePaths)):
  img = keras.preprocessing.image.load_img(imagePaths[i], target_size=(224, 224))
  x = keras.preprocessing.image.img_to_array(img)
  x = np.expand_dims(x, axis=0)
  x = keras.applications.resnet50.preprocess_input(x)
  predictions[i] = "Picasso" if model.predict(x) < 0.5 else "Rubens"
```

## show results

```python
for i in range(len(imagePaths)):
  img = mpimg.imread(imagePaths[i])
  imgplot = plt.imshow(img)
  plt.title(predictions[i])
  plt.show()
```
